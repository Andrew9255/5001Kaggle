```python
def preprocessing(data):
    #exttdate(data,'purchase_date')

    #exttdate(data,'release_date')
    
    data['purchase_weekday_median'] = data['purchase_date_weekday'].apply(lambda x: train_week_mean[x])
#     data['purchase_year_median'] = data['purchase_date_year'].apply(lambda x: train_year_median[x])
#     data['purchase_month_median'] = data['purchase_date_month'].apply(lambda x: train_month_median[x])
    
    data['purchase_weekday_median'].fillna(data['purchase_weekday_median'].mode()[0], inplace=True)
#     data['purchase_year_median'].fillna(data['purchase_year_median'].mode()[0], inplace=True)
#     data['purchase_month_median'].fillna(data['purchase_year_median'].mode()[0], inplace=True)
    
   
    data['datediff'] = (pd.to_datetime(data['purchase_date'])-pd.to_datetime(data['release_date'])).dt.days
    data['prepurchase'] = ((pd.to_datetime(data['purchase_date'])-pd.to_datetime(data['release_date'])).dt.days <=0)
    #Train['ifplay']= (Train['playtime_forever']>0)
    data['hot']=(data['total_positive_reviews']+data['total_negative_reviews'])
    data['hotm']=(data['hot'])/((2019-data['release_date_year'])+data['release_date_month']/12)
    #data["owndays"] = (pd.to_datetime('Nov 25, 2019')- pd.to_datetime(data['purchase_date'])).dt.days
    data['hotm'].fillna(data['hotm'].median(), inplace=True)
    data['purchase_date_year'].fillna(data['purchase_date_year'].median(), inplace=True)
    data['purchase_date_month'].fillna(data['purchase_date_month'].median(), inplace=True)
    data['purchase_date_weekday'].fillna(data['purchase_date_weekday'].median(), inplace=True)
    data['datediff'].fillna(data['datediff'].median(), inplace=True)
    #data['owndays'].fillna(data['owndays'].median(), inplace=True)
    
    #dumgenres =data['genres'].str.get_dummies(",").add_prefix("genres_")
    #dumgenres=dumgenres[['genres_Action','genres_Simulation']]
    
#     data['gen_count'] = data['genres'].apply(lambda x: len(x.split(',')))
#     data['cat_count'] = data['genres'].apply(lambda x: len(x.split(',')))
#     data['tag_count'] = data['genres'].apply(lambda x: len(x.split(',')))
#     data['tcount']=data['tag_count']+data['cat_count']+data['gen_count']
    
    #dumcategories=data['categories'].str.get_dummies(",").add_prefix("categories_")
    #dumcategories=dumcategories[['categories_Cross-Platform Multiplayer','categories_Online Multi-Player']]
    #dumcategories=dumcategories[['categories_Includes level editor','categories_Cross-Platform Multiplayer', 'categories_Steam Workshop','categories_Co-op','categories_Online Multi-Player']]
    dumtags=data['tags'].str.get_dummies(",").add_prefix("tags_")
    #dumtags=dumtags[['tags_Online Co-Op','tags_Crafting','tags_Open World','tags_PvP','tags_Rogue-like','tags_Survival','tags_Touch-Friendly','tags_Turn-Based Combat','tags_Turn-Based Tactics','tags_Difficult']]
    dumtags=dumtags[['tags_Online Co-Op','tags_Crafting','tags_PvP',]]
    #data=pd.concat([data,dumgenres],axis=1)
    #data=pd.concat([data,dumcategories],axis=1)
    data=pd.concat([data,dumtags],axis=1)

    
    
    return data

    
```

```python
def clean(data):
    data=data.drop(['genres','categories','tags'],axis=1)
    data=data.drop(['release_date_year','release_date_month','release_date_weekday'],axis=1)
    data=data.drop(['purchase_date_year','purchase_date_month','purchase_date_weekday'],axis=1)
    
    data=data.drop(['total_positive_reviews','total_negative_reviews','is_free','datediff','hot','price','prepurchase'],axis=1)
    #data=data.drop(['purchase_year_median','purchase_month_median'],axis=1)
    data =data.drop(['purchase_date','release_date',],axis = 1)
    
    return data
```

```python
train=preprocessing(train)
train=clean(train)

alldata=preprocessing(alldata)
alldata=clean(alldata)

null_columns=alldata.columns[alldata.isnull().any()]
alldata[null_columns].isnull().sum()
```

```python
def scaling_data(data):
    min_max_scaler = MinMaxScaler()
    data['hotm'] = min_max_scaler.fit_transform(np.array(data['hotm']).reshape(-1,1))
    data['tags_Online Co-Op'] = min_max_scaler.fit_transform(np.array(data['tags_Online Co-Op']).reshape(-1,1))
    data['tags_Crafting'] = min_max_scaler.fit_transform(np.array(data['tags_Crafting']).reshape(-1,1))
    data['tags_PvP'] = min_max_scaler.fit_transform(np.array(data['tags_PvP']).reshape(-1,1))
    data['purchase_weekday_median'] = min_max_scaler.fit_transform(np.array(data['purchase_weekday_median']).reshape(-1,1))
    return data
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from math import sqrt

forest_model = RandomForestRegressor(n_estimators=80, criterion='mae', oob_score=True, random_state=1, max_depth = 2)
forest_model.fit(train_X, train_y)
forest_rough_preds = forest_model.predict(train_X)
forest_rough_preds[forest_rough_preds<0]=0
print(sqrt(mean_squared_error(train_y, forest_rough_preds)))
forest_rough_preds = forest_model.predict(val_X)
forest_rough_preds[forest_rough_preds<0]=0
print(sqrt(mean_squared_error(val_y, forest_rough_preds)))
```
